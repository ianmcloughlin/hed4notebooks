## What is a Jupyter notebook?

***

A Jupyter notebook is like an interactive Microsoft Word document.

Programming code can be intermingled with text, graphics, and mathematical symbols.

The idea is to use the code to demonstrate concepts and ideas.

The code is typically written in the Python programming language.

```python
# Some example code.
# Print the first few powers of 3.
# We will write 3^2 when we mean 3 to the power of 2, etc.
# If you look at the code, Python uses the notation 3**2 instead.

for i in range(10):
    print(f"3^{i} = {3**i}")
```

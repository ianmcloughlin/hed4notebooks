## Sum of Waves

***

```python
# A range of x values for plotting functions.
x = np.linspace(-np.pi, 6.0 * np.pi, 1000)

# Waves.
ys = np.vstack([
    np.sin(x),
    3.0 * np.sin(x) + 1.0,
    np.sin(x + 1.0)**2,
    np.sin(2.0 * x),
])
```

```python
# Create new plot.
fig, ax = plt.subplots(figsize=(20, 8))

for y in ys:
    # Plot component wave.
    ax.plot(x, y, color='lightslategrey', linestyle=':')

# Plot sum wave.
ax.plot(x, ys.sum(axis=0), color='deeppink', label=r'$\Sigma$')

# Draw vertical lines at -pi and pi.
plt.axvline(x=0, color='teal', linestyle='--')
plt.axvline(x=2*np.pi, color='teal', linestyle='--')

# Show legend.
ax.legend();
```

## Waves with Similar Wavelengths

***

```python
# Factor.
f = 0.95

# A range of x values for plotting functions.
x = np.linspace(-60.0 * np.pi, 60.0 * np.pi, 1000)

# Sin wave.
y_sinA = np.sin(x)

# Sin wave.
y_sinB = np.sin(x * f)

# Sum.
y_sum = y_sinA + y_sinB
```

```python
# Create new plot.
fig, ax = plt.subplots(figsize=(20, 8))

# Plot Sin wave.
# ax.plot(x, y_sinA, color='teal', label='$\sin(x)$', linestyle=':')

# Plot Sin+2 wave.
# ax.plot(x, y_sinB, color='chocolate', label=f'$\sin(x * {f})$', linestyle=':')

# Plot the sum.
ax.plot(x, y_sum, color='mediumvioletred', label=fr'$\sin(x) + \sin({f} x)$')


# Show legend.
ax.legend();
```

```python
(np.exp(-1.0j * 2.0 * np.pi / len(y_sum) * np.arange(len(y_sum))))
```

```python
# Discrete Fourier Transform.
y_dft = np.zeros(len(y_sum), dtype=np.clongdouble)
for k in range(len(y_sum)):
    y_dft[k] = np.sum(y_sum * np.exp(-1.0j * 2.0 * np.pi * k / len(y_sum) * np.arange(len(y_sum))))
y_dft
```

```python
# Create new plot.
fig, ax = plt.subplots(figsize=(20, 8))

# Plot the sum.
ax.plot(x, y_dft.real, color='mediumvioletred')
```

```python
fig = plt.figure(figsize=(10, 10))

ax = fig.add_subplot(projection='3d')

ax.plot3D(x, y_dft.real, y_dft.imag);
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(20, 8))
ax = fig.add_subplot(projection='3d')

ax.contour3D(X, Y, Z, 800, cmap='winter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

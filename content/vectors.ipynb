```python
import numpy as np

import matplotlib.pyplot as plt
```

## Div, Grad, Curl, and All that (Schey)

```python
N = 10
x, y = np.meshgrid(np.arange(-N, N), np.arange(-N, N))
```

```python
# Problem 1.1 (a): iy + jx
i, j = y, x

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=100, angles='xy');
```

```python
# Problem 1.1 (b): (i + j)/sqrt(2)
i, j = 0 * x + 1.0 / np.sqrt(2), 0 * y + 1.0 / np.sqrt(2)

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=40, angles='xy');
```

```python
# Problem 1.1 (c): ix - jy
i, j = x, -y

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=100, angles='xy');
```

```python
# Problem 1.1 (d): iy
i, j = y, 0 * x

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=200, angles='xy');
```

```python
# Problem 1.1 (e): jx
i, j = 0 * x, x

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=200, angles='xy');
```

```python
# Problem 1.1 (f): (iy + jx)/sqrt(x^2 + y^2)
mag = np.sqrt(x**2 + y**2)
i, j = y / mag, x / mag

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=50, angles='xy');
```

```python
# Problem 1.1 (g): iy + jxy
i, j = y, x * y

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=500, angles='xy');
```

```python
# Problem 1.1 (h): i + jy
i, j = 0.0 * x + 1.0, y

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=50, angles='xy');
```

```python
# Problem 1.2
e_0 = 8.854 #e-12
E = np.divide(1.0 / (4.0 * np.pi *e_0), np.sqrt(x**2 + y**2), out=np.zeros_like(x, dtype=float), where=np.sqrt(x**2 + y**2)!=0) # 4.0 * np.pi * e_0 

i, j = -np.sign(x) * E, -np.sign(y) * E

i[N//2, N//2] = 0.0
j[N//2, N//2] = 0.0

# Plot.
fig, ax = plt.subplots(figsize=(10,10))

ax.quiver(x, y, i, j, scale=0.2, angles='xy');
```

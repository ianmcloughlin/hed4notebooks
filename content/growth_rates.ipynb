## Linear

***

Linear relationships in one variable are of the form $y = mx + c$.

The value $m$ is often called the slope of the line.

It is the change in $y$ due to a change in $x$.

The value $c$ is usually called the $y$-intercept.

It is the value of $y$ when $x$ is zero.

This means $(0,c)$ is the point where the line hits the $y$ axis.

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# These three lines with the same m value are parallel.
ax.plot(x, 2.0 * x + 1.0, 'b', label=r'$2x + 1$')
ax.plot(x, 2.0 * x + 4.0, 'g', label=r'$2x + 4$')
ax.plot(x, 2.0 * x + 9.0, 'r', label=r'$2x + 9$')

# Plot the y-axis.
ax.axvline(0.0, color='grey')

# Show the legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# These three lines with the same c value hit the y axis at the same spot.
ax.plot(x, -2.0 * x + 1.0, 'b', label=r'$-2x + 1$')
ax.plot(x,  1.0 * x + 1.0, 'g', label=r'$1x + 1$')
ax.plot(x,  2.0 * x + 1.0, 'r', label=r'$2x + 1$')

# Plot the y-axis.
ax.axvline(0.0, color='grey')

# Show the legend.
ax.legend();
```

## Polynomial

***

Expressions involving sums of fixed powers of $x$ multiplied by constants are called polynomials.

The following are examples: $x^2 + 1$, $4x^3 + 3x^2 + 2x + 1$, and $2x + 1$.

Notice that straight lines like the above represent polynomials too.

The highest power of $x$ in a polynomial is its degree.

When the highhest power is $2$, the polynomial is called a quadratic.

When it is $3$, it is called a cubic.

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, x**2, label=r'$y = x^2$')

# Legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, x**2, label=r'$y = x^2$')
ax.plot(x, x**2 - 10, label=r'$y = x^2 - 10$')
ax.plot(x, x**2 + 10, label=r'$y = x^2 + 10$')

# Legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, -x**2, label=r'$y = -x^2$')

# Legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, x**3, label=r'$y = x^3$')
ax.plot(x, x**3 - x**2 + 1, label=r'$y = x^3 - x^2 + 1$')
ax.plot(x, x**3 - 4*x**2 + 5*x - 10, label=r'$y = x^3 - 4x^2 + 5x - 10$')

# Legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, x**4, label=r'$y = x^4$')

# Legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, x**5, label=r'$y = x^5$')

# Legend.
ax.legend();
```

## Exponential

***

Exponentials involve $x$ in powers.

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, 2**x, label=r'$y = 2^x$')

# Legend.
ax.legend();
```

```python
# Create a plot.
fig, ax = plt.subplots(figsize=(12,6))

# Plot polynomial.
ax.plot(x, 2**x, label=r'$y = 2^x$')
ax.plot(x, x**2, label=r'$y = x^2$')

# Legend.
ax.legend();
```

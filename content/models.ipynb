```python
def square(x):
  return x * x
```

```python
square(3)
```

```python
square(10)
```

```python
def power(x, n):
  y = x
  for i in range(1, n):
    y = y * x
  return y    
```

```python
power(3, 2)
```

```python
power(3, 4)
```

```python
power(10, 10)
```

<h2 style="color: rgb(0, 91, 94);">From Data</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

```python
import numpy as np
```

```python
no_points = 1000
```

```python
x = np.sort(np.random.uniform(1.0, 10.0, no_points))
x[:10]
```

```python
secret = 4
```

```python
y = x**secret
```

```python
y = y + np.random.normal(0.0, 100.0, no_points)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, '.')
plt.show();
```

```python
plt.plot(x, y, '.')
x_sorted = np.sort(x)
plt.plot(x_sorted, x_sorted**secret, label=f'$x^{secret}$')
plt.plot(x_sorted, x_sorted**(secret - 1), label=f'$x^{secret - 1}$')
plt.legend()
plt.show();
```

$ y = x^n $

$ \Rightarrow \ln y = \ln x^n $

$ \Rightarrow \ln y = n \ln x $

```python
ln_x = np.log(x)
ln_y = np.log(y)
```

```python
ln_x = ln_x[~np.isnan(ln_y)]
ln_y = ln_y[~np.isnan(ln_y)]
```

```python
m, c = np.polyfit(ln_x, ln_y, 1)
m, c
```

```python
best_m = np.sum(ln_x * ln_y) / np.sum(ln_x * ln_x)
best_m
```

```python
plt.plot(ln_x, ln_y, '.', label='Data')
plt.plot(ln_x, m * ln_x + c, label='polyfit')
plt.plot(ln_x, m * ln_x, label='calculation')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, '.', label='Data')
plt.plot(x, c * x**m, label='polyfit')
plt.plot(x, x**best_m, label='calculation')
plt.legend()
plt.show()
```

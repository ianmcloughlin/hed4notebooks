## Counting

***

We use numbers in lots of different situations where they serve different purposes.

We can use them to count - for which we usually use the natural numbers:

$\mathbb{N} = \{ 1, 2, 3, \ldots\}$

Sometimes we include zero as a counting number:

$\mathbb{N}_0 = \{ 0, 1, 2, 3, \ldots\}$

And sometimes we even count with negative numbers - for example when keeping track of money someone owes us.

Then we say we are using the *integers*:

$\mathbb{Z} = \{ \ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\}$

The integers are so named because they stand all by themselves.

They do not touch each other on the number line.

## Measuring

***

We also use numbers for measuring.

How high is the ceiling?

How far away is the sun?

The measuring numbers are qualitatively different than the counting numbers.

We usually depict them on a number line:

![The Number Line](https://upload.wikimedia.org/wikipedia/commons/thumb/d/d7/Real_number_line.svg/320px-Real_number_line.svg.png)

They are called the real numbers $\mathbb{R}$.

We cannot list them out like we do the integers, because between any two distinct real numbers is another real number.

For example, between 0.0001 and 0.0002 is 0.00005.

Of course, there is also 0.00001, 0.000001, 0.00000000001.

In fact, there is an infinity of numbers between any two distinct real numbers.

That is why we use them for measuring - we can be as accurate as we like.

## Solving Equations

***

A common maths exercise is to set some expression in $x$ to $0$ and figure out what $x$ should be.

For example, take $f(x) = x^2 - 4x + 3$.

Setting $f(x) = 0$ gives $x^2 - 4x + 3 = 0$.

Let us plot $f(x)$ and see if we can see what $x$ might be.

## Setup

***

We will use some common Python libraries to create the plot.

```python
# Numerical arrays.
import numpy as np

# Plotting.
import matplotlib.pyplot as plt

# Set a better plot size.
plt.rcParams["figure.figsize"] = (16, 8)
```

```python
# Create some x values.
x = np.linspace(-1.0, 5.0, 100)

# Set the corresponding y values.
y = x**2 - 4 * x + 3
```

```python
# Create the plot.
plt.plot(x, y)

# Show the x-axis in pink;
plt.axhline(0.0, color=(0.95, 0.5, 0.9));
```

It looks like the plot of $f(x)=x^2 - 4x + 3$ crosses the x-axis at $1$ and $3$.

We can check that: $f(1) = 1^2 - 4(1) + 3 = 1 - 4 + 3 = 0$ and $f(3) = 3^2 - 4(3) + 3 = 9 - 12 + 3 = 0$.

So $f(x)$ is 0 when $x$ is $1$ and $x$ is $3$.

It can be shown that there are at most two such distinct values for an expression like $f(x)$.

There is an important facet of this: we do not need anything but integers for $f(x)$ and the values of $x$.

## Another Example

***

What about $f(x) =  2x^2 - 3x + 1$?

Again, let's plot it.

```python
# Create some x values.
x = np.linspace(0.0, 1.5, 100)

# Set the corresponding y values.
y = 2 * x**2 - 3 * x + 1

# Create the plot.
plt.plot(x, y)

# Show the x-axis in pink;
plt.axhline(0.0, color=(0.95, 0.5, 0.9));
```

Now it looks like the right-most value of $x$ is $1$.

Let's confirm it: $f(1) = 2(1)^2 - 3(1) + 1 = 2 - 3 + 1 = 0$.

What about the left-most value?

Let's try $0.5$: $f(0.5) = 2 (0.5)^2 - 3(0.5) + 1 = 0.5 - 1.5 + 1 = 0$.

There's an import point here: $0.5$ is not an integer.

So sometimes we need more than integers.

## One Last Example

***

Now let us try a seemingly simple looking expression, $f(x) = x^2 + 1$.

What values of $x$ make $f(x) zero?

Let's plot.

```python
# Create some x values.
x = np.linspace(-2.0, 2.0, 100)

# Set the corresponding y values.
y = 2 * x**2 + 1

# Create the plot.
plt.plot(x, y)

# Show the x-axis in pink;
plt.axhline(0.0, color=(0.95, 0.5, 0.9));
```

There is a real problem here (pardon the pun).

The blue curve does not cross the pink line.

It looks like there is no solution.

However, there is a formula that always works - many students learn it in school.

When $f(x) = ax^2 + bx + c = 0$ then:

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

Let's plug our values in: $a$ is $1$, $b$ is $0$, and $c$ is $1$.

So, $x = \frac{\pm \sqrt{-4}}{2} = \frac{\pm \sqrt{4} \sqrt{-1}}{2} = \pm \frac{2}{2} \sqrt{-1} = \pm \sqrt{-1}$.

What do we make of $\sqrt{-1}$? What can we say about it?

First of all, it is something that when we mutliply it by itself it gives $-1$.

Well, no real number has this property because whenever you multiply a real number by itself it gives a positive answer.

For example, $2 \times 2 = 4$ and $-2 \times -2 = 4$.

Maybe we can't do anything with it.

It does work, though: $f(\sqrt{-1}) = (\sqrt{-1})^2 + 1 = -1 + 1 = 0$.

## $i$

The square root of minus one is just itself.

There is no integer or real number it equals.

It is not on the real nubmer line.

Sometimes people call it imaginary, but it isn't.

It's just necessary to solve some expressions in $x$.

It has very real uses.

It is so important it is built into Python.

Where most books use the easier to write $i$ for $\sqrt{-1}$, Python uses $j$.

## Built-in

***

You can use the square root of minus one as follows in Python.

```python
1j
```

You have to put the 1 in front of it.

Let's multiply it by itself.

```python
1j * 1j
```

It gives the answer $-1 + 0j$.

It is quite common to write numbers involving $i$ or $j$ as $a + bi$ or $a + bj$.

The values $a$ and $b$ being real numbers.

Numbers like these are called complex numbers and the set of all of them is denoted $\mathbb{C}$.

```python
-2.0 + 5.0j
```

We can add two complex numbers.

```python
(5 + 2j) + (-1 - 9j)
```

We can raise them to powers.

```python
(5 + 2j)**2
```

We can multiply them too.

```python
(5 + 2j) * (-1 - 9j)
```

## As Pairs

***

In many ways, we can just think of complex numbers as pairs of numbers (a, b).

Sometimes these are called tuples.

Adding tuples is just done component-wise:

$(a,b) + (c,d) = (a+c,b+d)$

```python
x = (-2.0, 5.0)
x
```

```python
y = (3.0, 1.5)
y
```

```python
def add(u, v):
    return (u[0] + v[0], u[1] + v[1])
```

```python
add(x, y)
```

This is just the same as adding the corresponding complex numbers.

```python
(-2 + 5j) + (3 + 1.5j)
```

Multiplication is a little trickier.

$(a,b) \times (c,d) = (ac - bd, ad + bc)$

```python
def mult(u, v):
    a, b = u
    c, d = v
    return ((a * c) - (b * d), (b * c) + (a * d))
```

```python
mult(x, y)
```

Again, this is the same as multiplying the corresponding complex numbers.

```python
(-2 + 5j) * (3 + 1.5j)
```
